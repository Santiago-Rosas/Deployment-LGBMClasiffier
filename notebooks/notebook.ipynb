# Data Trasnformation

```python
##change SMK_stat_type_cd from float to  categorical 
df_2["SMK_stat_type_cd"]=df_2["SMK_stat_type_cd"].astype('category')
```

```python
# transform all object  variables to numerical 
column_equivalence = {}
for i, column in enumerate(df_2.dtypes):
    if column == "object":
        categorical_column = df_2[df_2.columns[i]].astype("category")
        current_column_equivalence = dict(enumerate(categorical_column.cat.categories))
        column_equivalence[i] = dict((v,k) for k,v in current_column_equivalence.items())
        df_2[df_2.columns[i]] = categorical_column.cat.codes
```

## Subsampling 

```python
##  The data is balanced but we need to subsample due to computer resources.
X= df_2.drop(columns="DRK_YN")
y=df_2["DRK_YN"]
```

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(random_state=42,sampling_strategy={0:20000,1:20000})
```

```python
y
```

```python
X_over , y_over = undersample.fit_resample(X,y)
```

```python
X_over.shape
```

```python
y_over.shape
```

```python
y_plot=pd.DataFrame(y_over)
sns.countplot(data=y_plot,x="DRK_YN", palette="Set2")
plt.show()
```

## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X_over, y_over, test_size=0.33, random_state=42)
```

## Data scaling 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
##make a list of the index columns  that we need to escale 
columns_scale=[]
not_scale=[]
for index, column in enumerate(X.columns):
    if len(df_2[column].unique()) > 5 and X[column].max() > 2:
        columns_scale.append(index)
        print(column, f"number of unique values: {len(X[column].unique())}")
    else:
        not_scale.append(index)
print("index to sacle:",columns_scale)
print("not sacle:",not_scale)
```

```python
scaler = ColumnTransformer(
[("scale",StandardScaler(),columns_scale)],remainder= 'passthrough'
)
```

```python
scaler.fit(X_train)
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

# Tunning Hyperparameters with GridSearchCV
 we could see that this models are good options:
- SVC
- GradientBoostingClassifier
- LogisticRegression
- RandomForestClassifier
- LGBMClassifier

```python
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMClassifier
####metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
reg = {
        'SVC' : SVC(),
        'GRADIENT' : GradientBoostingClassifier(),
        'LR': LogisticRegression(),
        'RF':RandomForestClassifier(),
        'lGBM': LGBMClassifier()
         }

params = {
    'SVC' : {
        'kernel' : ['linear', 'poly', 'rbf'],
        'gamma' : ['auto', 'scale'],
        'C' : [0.001, 0.01,0.1,1,1.5,2],
        
        },

    'GRADIENT' : {
        'learning_rate' : [0.001,0.01, 0.05,0.1,0.2,0.3,0.4],
        'n_estimators':[20,50,100,200]
           },

    'LR' :{
             'solver': ['lbfgs','newton-cg', 'sag', 'saga'],
             'C' : [0.2,1,1.5,2,10],
             
             
           },

    'RF': {
          'n_estimators': [20,50,100,200,300],
          'criterion': ['gini', 'entropy', 'log_loss'],
          
          },
          
    'lGBM':{
          "max_depth": list(range(3,10)),
          "learning_rate":[0.001, 0.01,0.1],
          "num_iterations":[300],
          "num_leaves":[5,10,15,31,40,50,60,100]
        
          
        }
         
        }
```

```python
best_model_SVC=None
best_model_GBC=None
best_model_LR=None
best_model_RF=None
best_model_LGBM=None
models_best_score = dict() ##dictionary with the name of the best models and tehir scores 

for name, reg in reg.items():

  

    grid_reg = GridSearchCV(reg, params[name], cv=3).fit(X_train, y_train.values.ravel())
    score = np.abs(grid_reg.best_score_)
    print(name,grid_reg.best_params_)

    if score >= 0.60:
        models_best_score.update({name:score})

    if name == "SVC":
        best_model_SVC= grid_reg.best_estimator_
        
    if name == "GRADIENT":
        best_model_GBC= grid_reg.best_estimator_

    if name == 'LR':
        best_model_LR= grid_reg.best_estimator_ 
    
    if name == 'RF':
        best_model_RF= grid_reg.best_estimator_ 

    else:

        best_model_LGBM= grid_reg.best_estimator_ 
```

```python
models_best_score
```

```python
mejores_modelos= [best_model_LGBM, best_model_SVC,best_model_GBC, best_model_LR,best_model_RF]

for i,mod in  enumerate(mejores_modelos):
    print("***"*25)
    print(f"modelo {i+1}")
    print("***"*25)
    print(mod)
    print(f"accuracy score:{accuracy_score(y_test,mod.predict(X_test))}")
    print(f"f1_score:{f1_score(y_test,mod.predict(X_test))}") 
    print("confusion matrix")
    print(confusion_matrix(y_test,mod.predict(X_test)))
    
```
